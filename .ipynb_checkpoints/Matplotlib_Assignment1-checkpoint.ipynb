```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
#my list
listGPA=[1, 1.5, 2, 2.5, 3, 3.5, 4, 4.4, 4.5, 4.6]
#my dataframe
dataframe=pd.DataFrame([['Fai',4.4],['Mai',4.5],['Maha', 4.6]], columns=['Names', 'GPA'])

#plotting the list and dataframe
plt.title('list and dataframe',fontsize = 20)
plt.plot(listGPA, 'r')
plt.plot(dataframe['GPA'],'b')

plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(dataframe['GPA'],labels=dataframe['Names'], autopct='%1.1f%%', startangle=90, shadow=True, radius=1, explode = [0.2,0.2,0.2])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(dataframe['Names'],[dataframe['GPA']], alpha=0.4, marker = 'o', s = 50, label = 'Students GPA');
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(dataframe['Names'],dataframe['GPA'], alpha=0.4, marker = 'o', s = 50, label = 'Students GPA')
plt.title('Students GPA',c="r", loc='left')
plt.xlabel('Names',fontsize = 16)
plt.ylabel('GPA',fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
#list 1
plt.plot([1,2,3,4,5], 'r')
#list 2
plt.plot([6,7,8,9,10],'b')
#list 3
plt.plot([10, 8, 6, 4, 2], 'g')

plt.legend(['list1','list2', 'list3'], shadow = True, loc = (6))
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python

plt.figure(figsize=[20,10])
plt.suptitle('Three Charts',fontsize = 20)
#bar graph plot
plt.subplot(2,2,1) 
plt.bar(dataframe['Names'], dataframe['GPA'])

#pie chart plot
plt.subplot(2,2,2) 
plt.pie(dataframe['GPA'],labels=dataframe['Names'], autopct='%1.1f%%', startangle=90, shadow=True, radius=1)

#histogram plot
plt.subplot(2,2,3) 
#took students GPA as x because histogram gives better representation with numbers (two students have GPA from 4.5 to 5.0)
plt.hist(dataframe['GPA'], bins=[1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]);
plt.show()
```
